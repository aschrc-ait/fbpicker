### DATA FORMAT
Setting up data containers<br><br>

The whole functionality here is based on the Hierarchical Data Format (HDF) version 5.<br>
It is a data model, library, and open file format for storing and managing Terabyte scale complex scientific data.<br>
It supports an unlimited variety of data types, and is designed for flexible and efficient I/O of high volume data.<br>
H5py, a Pythonic interface to the HDF5 binary data format, uses straightforward NumPy and Python metaphors, like dictionary and NumPy array syntax.<br>
For example, you can iterate over datasets in a file, or check out the .shape or .dtype attributes of datasets.<br><br>
In case of seismic data storage, HDF5 offers many improvements over SEG-Y, including:
- better bindings and interchangeability between different applications and languages
- seamless handling of endian issues
- support for chunked multidimensional datasets
- supports for parallel I/O

The exemplary seismic dataset used below was exported to .hdf5 format with Claritas processing software.<br>
All Claritas HDF5 files have the trace headers stored in a HDF5 table, and trace data and pseudotraces as HDF5 datasets.<br>
These are both stored under the group /TRACE_DATA/DEFAULT to allow for storing any number of seismic data collections in a single file.<br>
The trace headers objects (HDF5 table) are equivalent to all of the trace headers stored in a SEG-Y file.<br>
Unlike SEG-Y, these headers are not interleaved with the data. Each field or column in this table corresponds to a named trace header,<br>
and each row of the table corresponds to a trace in the file.<br>
The 'data_array' object is an extendible 2D dataset of native 32 bit IEEE floating point data, extendible in both dimensions.<br>
Endian issues are automatically handled if reading a file written on a system with a different endian-ness.<br>
Each column in this dataset corresponds to a sample value in the trace, and each row corresponds to a trace.<br><br>

Loading first-break models and feature lists

```python
m_1s=load_model('./models/m_1s.h5')
m_50s_prefb=load_model('./models/m_50s_prefb.h5')
m_50s_postfb=load_model('./models/m_50s_postfb.h5')
models=np.hstack((m_1s,m_50s_prefb,m_50s_postfb))
features_prefb=np.load('./models/best_features_prefb.npy')
features_postfb=np.load('./models/best_features_postfb.npy')
```

Setting few starting parameters
- dt - sampling interval [ms]
- min_offset/max_offset - minimal and maximal offset [m]
- scalar_offset - scalar to be applied to spatial coordinates<br>
(e.g. given real value 1000.5 and stored value 10005, the scalar value is 10)
- first_sample - time sample where fbpicker starts
- trc_slen - number of time samples that fbpicker analyzes

```python
dt=2
min_offset=0
max_offset=3500
scalar_offset=10
first_sample=0
trc_slen=500
```

Creating a instance of fbpicker class

```python
fbpicker=fbpicker(d1,models,min_offset,max_offset,scalar_offset,first_sample,trc_slen,dt,features_prefb,features_postfb)
```

Performing automatic first-break detection

```python
fbpicker.predict()
```

Deriving first-break time samples by finding the first nth percentile of probability distribution within each trace

```python
fbs=fbpicker.find_fbs(q=99.99)
```

Generating basic statistics on automatic first-break picking outcome

```python
fbpicker.stats(allowed_diff=1,predicted_fbs=fbs,reference_fbs=d1['gapsize'][:]/2)
```
